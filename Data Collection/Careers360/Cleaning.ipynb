```python
import pandas as pd

# --- CONFIGURATION ---
INPUT_FILE = "careers360_raw_reviews.csv"  # The file you want to clean
OUTPUT_FILE = "reviews_cleaned_names.csv"  # The new file to save

def clean_college_names():
    try:
        print(f"📂 Loading {INPUT_FILE}...")
        df = pd.read_csv(INPUT_FILE)

        # Check if column exists
        if 'college_name' not in df.columns:
            print("❌ Error: 'college_name' column not found in the CSV.")
            return

        print("🧹 Cleaning college names...")

        # Remove ',' and '.' using regex
        # regex=True allows us to use '[.,]' which matches either a dot or a comma
        df['college_name'] = df['college_name'].str.replace(r'[.,]', '', regex=True)

        # Optional: Remove extra spaces that might be left behind (e.g., "City , State" -> "City  State")
        df['college_name'] = df['college_name'].str.replace(r'\s+', ' ', regex=True).str.strip()

        # Save to new file
        df.to_csv(OUTPUT_FILE, index=False)
        
        print(f"🎉 Done! cleaned data saved to '{OUTPUT_FILE}'")
        print("\n--- Preview of cleaned names ---")
        print(df['college_name'].head())

    except FileNotFoundError:
        print(f"❌ File '{INPUT_FILE}' not found. Please check the name.")
    except Exception as e:
        print(f"❌ An error occurred: {e}")

if __name__ == "__main__":
    clean_college_names()
```

```python
import pandas as pd

# --- CONFIGURATION ---
INPUT_FILE = "reviews_cleaned_names.csv"   # Your cleaned CSV file
OUTPUT_TXT_FILE = "unique_colleges.txt"    # The text file to be created

def save_unique_names():
    try:
        print(f"📂 Loading {INPUT_FILE}...")
        df = pd.read_csv(INPUT_FILE)

        # 1. Get unique names and sort them alphabetically
        unique_colleges = sorted(df['college_name'].unique())

        # 2. Save to a text file
        with open(OUTPUT_TXT_FILE, "w", encoding="utf-8") as f:
            for name in unique_colleges:
                f.write(f"{name}\n")

        print(f"✅ Success! {len(unique_colleges)} unique names saved to '{OUTPUT_TXT_FILE}'")

    except FileNotFoundError:
        print(f"❌ File '{INPUT_FILE}' not found.")
    except Exception as e:
        print(f"❌ Error: {e}")

if __name__ == "__main__":
    save_unique_names()
```

```python
# --- CONFIGURATION ---
FILE_1 = "college_list.txt"      # Your list of desired colleges
FILE_2 = "unique_colleges.txt"   # The list found in the dataset

def normalize(text):
    """Helper to lowercase and strip spaces for comparison"""
    return text.strip().lower()

def compare_files():
    try:
        # 1. Read File 1 (Preserve original names for printing)
        with open(FILE_1, "r", encoding="utf-8") as f:
            list_1 = [line.strip() for line in f if line.strip()]

        # 2. Read File 2 (Convert to a set for fast lookup)
        with open(FILE_2, "r", encoding="utf-8") as f:
            # We only need the normalized version for checking
            set_2 = set(normalize(line) for line in f if line.strip())

        print(f"📂 Loaded {len(list_1)} names from {FILE_1}")
        print(f"📂 Loaded {len(set_2)} unique names from {FILE_2}")
        print("-" * 30)

        matches = 0
        missing_names = []

        # 3. Check for matches
        for name in list_1:
            if normalize(name) in set_2:
                matches += 1
            else:
                missing_names.append(name)

        # 4. Output Results
        print(f"✅ Found Matches: {matches}")
        print(f"❌ Not Found (Missing): {len(missing_names)}\n")

        if missing_names:
            print("--- Names in File 1 NOT found in File 2 ---")
            for missing in missing_names:
                print(f" • {missing}")
        else:
            print("🎉 Perfect Match! All names were found.")

    except FileNotFoundError as e:
        print(f"❌ Error: Could not find file. {e}")
    except Exception as e:
        print(f"❌ Error: {e}")

if __name__ == "__main__":
    compare_files()
```

```python
import pandas as pd

# Load files
college_list = set(
    line.strip()
    for line in open("college_list.txt", encoding="utf-8")
    if line.strip()
)

unique_colleges = set(
    line.strip()
    for line in open("unique_colleges.txt", encoding="utf-8")
    if line.strip()
)

map_df = pd.read_csv("map_final.csv")

missing_colleges = []

for college in college_list:
    # Find matching row in map.csv
    row = map_df[
        (map_df["college_name_1"] == college) |
        (map_df["college_name_2"] == college) |
        (map_df["college_name_3"] == college) |
        (map_df["college_name_4"] == college)
    ]

    if row.empty:
        # College not present in map.csv at all
        missing_colleges.append((college, "NOT FOUND IN MAP"))
        continue

    # Collect all possible mapped names
    mapped_names = set()
    for col in ["college_name_1", "college_name_2", "college_name_3", "college_name_4"]:
        val = row.iloc[0][col]
        if pd.notna(val):
            mapped_names.add(val.strip())

    # Check if any mapped name exists in unique_colleges
    if not any(name in unique_colleges for name in mapped_names):
        missing_colleges.append((college, "MAPPED BUT NOT IN DATASET"))

# Print results
print("\n❌ Colleges with missing mappings:\n")
for college, reason in missing_colleges:
    print(f"{college} --> {reason}")

print(f"\nTotal missing colleges: {len(missing_colleges)}")
```

```python
import pandas as pd

# Load files
reviews_df = pd.read_csv("reviews_cleaned_names.csv")
map_df = pd.read_csv("map_final.csv")

# Create lookup dictionary: college_name_variant -> college_id
name_to_id = {}

for _, row in map_df.iterrows():
    college_id = row["college_id"]
    for col in ["college_name_1", "college_name_2", "college_name_3", "college_name_4"]:
        name = row[col]
        if pd.notna(name):
            name_to_id[name.strip()] = college_id

# Function to map college_name to college_id
def get_college_id(college_name):
    return name_to_id.get(college_name, None)

# Apply mapping
reviews_df["college_id"] = reviews_df["college_name"].apply(get_college_id)

# Keep only matched reviews
final_df = reviews_df.dropna(subset=["college_id"]).copy()

# Convert college_id to int
final_df["college_id"] = final_df["college_id"].astype(int)

# Optional: reorder columns
final_df = final_df[
    [
        "college_id",
        "college_name",
        "review_title",
        "overall_rating",
        "review_text",
        "posted_info",
        "source",
        "page_number"
    ]
]

# Save final dataset
final_df.to_csv("careers360_final_dataset.csv", index=False)

print("✅ Final dataset created")
print(f"Total reviews kept: {len(final_df)}")
print(f"Unique colleges: {final_df['college_id'].nunique()}")
```

```python
import pandas as pd

df = pd.read_csv(
    "careers360_final_dataset.csv",
    encoding="latin1"   # or "ISO-8859-1"
)


review_counts = (
    df.groupby(["college_id"])
      .size()
      .reset_index(name="num_reviews")
      .sort_values("college_id", ascending=True)
)

print(review_counts)
review_counts.to_csv("college_review_counts.csv", index=False)
```
